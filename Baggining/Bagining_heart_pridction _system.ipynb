```python
# Bagging (Bootstrap Aggregating) is an ensemble learning technique that improves the stability and accuracy of machine learning algorithms.
# Key features of bagging:
# - It creates multiple subsets of the original dataset using random sampling with replacement (bootstrap samples).
# - Each subset is used to train a separate base model (often decision trees).
# - The predictions from all models are combined (averaged for regression, majority vote for classification) to produce the final output.
# - Bagging helps to reduce variance and prevent overfitting, especially for high-variance models.
# - It works well with algorithms that are sensitive to small changes in the training data.
# - RandomForest is a popular algorithm that uses bagging with decision trees and adds feature randomness for further improvement.
#also write how it wroks about bagging
# Bagging works by training multiple models on different subsets of the data and then combining their predictions.
# The process involves the following steps:
# 1. Create multiple bootstrap samples from the original dataset.
# 2. Train a separate base model on each bootstrap sample.
# 3. For classification tasks, use majority voting to combine the predictions from all models.
# 4. For regression tasks, average the predictions from all models.
# This approach helps to reduce overfitting and improve the overall performance of the model.


#write about varinace and bisas 
# Variance refers to the model's sensitivity to fluctuations in the training data. High variance models can overfit, capturing noise rather than the underlying pattern.
# Bias refers to the model's tendency to consistently predict a certain way, leading to systematic errors. High bias models can underfit, failing to capture the underlying pattern in the data.



```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df[df.MaxHR>(df.MaxHR.mean()+3*df.MaxHR.std())]
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
```

```python
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df.ChestPainType.unique()
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=20)
```

```python
X_train.shape
```

```python

from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(), X, y, cv=5)
scores.mean()
```

```python
#using Bagging with SVC for better performance
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    random_state=0
)
scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

```python
#using decision tree classifier  first 

from sklearn.tree import DecisionTreeClassifier

scores = cross_val_score(DecisionTreeClassifier(random_state=0), X, y, cv=5)
scores.mean()
```

```python
#using Bagging with Decision Tree Classifier for better performance 

bag_model = BaggingClassifier(
   estimator=DecisionTreeClassifier(random_state=0), 
    n_estimators=100, 
    max_samples=0.9, 
    oob_score=True,
    random_state=0
)

scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

```python
# Comments
#use RandomForestClassifier for better performance becase it is an ensemble method that combines multiple decision trees to improve accuracy and control overfitting.
#it has baging and feature randomness built-in, which helps in reducing overfitting and improving generalization.
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(), X, y, cv=5)
scores.mean()
```
