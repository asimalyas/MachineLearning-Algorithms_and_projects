```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import os


file_path = 'canada_per_capita_income.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(df.head())  # First 5 lines
else:
    print("File not found!")

model = linear_model.LinearRegression()

model.fit(df[['year']], df['per capita income (US$)'])

# Predict income for year 2020
predicted_income_2020 = model.predict([[2020]])
print(f"Predicted income for 2020: {predicted_income_2020[0]:.2f} USD")
p = model.predict([[2021]])
print("the estimated us dolorrs for : 2021 is : ",p)
```

```python
plt.scatter(df[["year"]] , model.predict(df[['year']]), label='Actual Data')
plt.xlabel("year")
plt.ylabel("per capita income")
```
